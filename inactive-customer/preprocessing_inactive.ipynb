**MÔ TẢ VẤN ĐỀ**

- Khởi đầu tất cả với 2 tệp là có khả năng rời bỏ / có khả năng Active trở lại
- Giờ đây bài toán mới yêu cầu chúng ta cần phải có nhiều trường hơn, và nhiều cụm hơn để phù hợp với nghiệp vụ hơn
- Các trường dữ liệu sử dụng:
    + AGE	Tuổi
    + NAV_BQ_12M: NAV bình quân 12 tháng (01/01/2024 - 31/12/2024)
    + TODAY_NAV: Nav hiện tại của họ (Thời điểm cụ thể lấy 31-12-2024)
    + NAV_LAST_TRADING: Nav tại ngày cuối cùng họ giao dịch 
    + CASH_IN: Tổng số tiền nạp vào từ thời điểm họ giao dịch lần cuối tới hiện tại 
    + CASH_OUT: Tổng số tiền rút ra từ thời điểm họ giao dịch lần cuối tới hiện tại 
    + VQGD_6M: Vòng quay giao dịch 6 tháng gần nhất trước thời điểm họ giao dịch lần cuối
    + TRADING_COUNT_6M: Số ngày giao dịch trong 6 tháng trước thời điểm họ giao dịch lần cuối
    + GTGD_6M: Giá trị giao dịch 6 tháng gần nhất trước thời điểm họ giao dịch lần cuối
    + PROFIT_LOSS_6M: Lãi lỗ thực hiện 6 tháng trước thời điểm họ giao dịch lần cuối
    + PROFIT_LOSS_PERCENTAGE_6M: Phần trăm lãi lỗ thực hiện 6 tháng trước thời điểm họ giao dịch lần cuối
    + TRADING_COUNT_1M: Số ngày giao dịch trong 1 tháng trước thời điểm họ giao dịch lần cuối
    + TRADING_COUNT_2M: Số ngày giao dịch trong 2 tháng trước thời điểm họ giao dịch lần cuối
    + TRADING_COUNT_3M: Số ngày giao dịch trong 3 tháng trước thời điểm họ giao dịch lần cuối
    + CHANNEL: Kênh mở
    + LOGIN_COUNT_12M: Số lần login 12 tháng năm nay (01/01/2024 - 31/12/2024)
    + LOGIN_COUNT_6M: Số lần Login trong 6 tháng gần đây (30/06/2024 - 31/12/2024)
    + LOGIN_COUNT_3M: Số lần Login trong 3 tháng gần đây (30/09/2024 - 31/12/2024)
    + ACTIVE_AGE: Khoảng thời gian Active từ lúc họ giao dịch lần đầu đến lúc thời điểm giao dịch lần cuối

**XỬ LÝ NULL**

- Từ đây ta xử lý các GT null như sau:
    + AGE: lấy mẹ trung vị: median, thay các giá trị âm bằng median
    + CASH_IN: thay = 0
    + CASH_OUT: thay = 0
    + VQGD_6M: chưa biết? Cho = 0 chăng? Hay là drop? ->> thay mẹ = 0
    + PROFIT_LOSS_6M: cũng chưa biết? má khó z ->>> thay mẹ = 0

    + LOGIN_COUNT_3M: Chắc chắn p thay = 0
    + LOGIN_COUNT_6M: Chắc chắn p thay = 0
    + LOGIN_COUNT_12M: Chắc chắn p thay = 0                 

```python
# thay các giá trị âm ở trường AGE bằng median
df.loc[df['AGE'] <= 0, 'AGE'] = df['AGE'].median()

preprocessing.fill_na_0(df, ['CASH_IN', 'CASH_OUT', 'LOGIN_COUNT_3M', 'LOGIN_COUNT_6M', 'LOGIN_COUNT_12M', 'VQGD_6M', 'PROFIT_LOSS_6M', 'PROFIT_LOSS_PERCENTAGE_6M'])

preprocessing.fill_na_median(df, ['AGE'])
```
